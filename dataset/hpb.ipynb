# Encoding categorical variables with multiple options

```python
multiple_options_columns = ['What are some features offered by these health tracking apps that you enjoy?',
                           'What difficulties do you face in staying motivated to use health tracking apps consistently and regularly?',
                            "What are the top 3 motivations for your participation in physical activities?",
                            "What types of rewards or incentives would encourage you to use an app regularly?",
                            "Select the top three gamification features that would be most appealing to you in a health tracking app",
                            "What features would make you more likely to participate in community events or challenges within the app?",
                            "How do you think schools could help promote health tracking apps (eg: H365) among students?"
                            ]

from sklearn.preprocessing import MultiLabelBinarizer

#df[multiple_options_columns] = df[multiple_options_columns].apply(lambda x: [item.strip() for item in x.split(',')])
```

```python
numbering = {'What are some features offered by these health tracking apps that you enjoy?': 'Q6', 
             'What difficulties do you face in staying motivated to use health tracking apps consistently and regularly?': 'Q8',
             "What are the top 3 motivations for your participation in physical activities?": 'Q14',
             "What types of rewards or incentives would encourage you to use an app regularly?": 'Q15',
             "Select the top three gamification features that would be most appealing to you in a health tracking app": 'Q18',
             "What features would make you more likely to participate in community events or challenges within the app?": 'Q21',
             "How do you think schools could help promote health tracking apps (eg: H365) among students?": 'Q24'
            }

def process_column(column):
    column = column.fillna('').apply(lambda x: [item.strip() for item in x.split(',')])
    mlb = MultiLabelBinarizer()
    X_encoded = mlb.fit_transform(column)
    names = [numbering[column.name] + '_' + name for name in mlb.classes_]
    return pd.DataFrame(X_encoded, columns=names)
```

```python
for col in multiple_options_columns:
    encoded_dfs = {}
    encoded_dfs[col] = process_column(df[col])
    df_encoded_multiple = pd.concat(encoded_dfs.values(), axis=1)
    print(col)
    print(df_encoded_multiple.columns)
```

```python
encoded_dfs = {}
for col in multiple_options_columns:
    encoded_dfs[col] = process_column(df[col])
```

```python
df_encoded_multiple = pd.concat(encoded_dfs.values(), axis=1)

comma_columns = ['Q6_profiles or app interfaces', 'Q6_view friends’ profiles', 'Q6_leaderboards)', 'Q6_achievement trophies','Q6_virtual currency)', 'Q14_weight management)']
df_encoded_multiple.drop(columns=comma_columns, inplace=True)

# others_columns = ['Q6_health app tracks period cycle', 'Q6_tracks the number of steps i take', 'Q8_the apps i use only support running', 'Q8_Lazy to start working out', 'Q14_Ice skating makes me feel pretty.', 'Q14_To make my physique look better', 'Q15_The thjngy where you pair with friendsss', 'Q15_cash voucher',
#                   'Q15_food and cash rewards', 'Q15_money', 'Q15_money hehe','Q15_money/vouchers', 'Q15_vouchers', 'Q18_exchange for money or vouchers', 'Q21_Monetary reward', 'Q24_provide vouchers/prizes for those that achieve more fitness']
#df_encoded_multiple.drop(columns=others_columns, inplace=True)

rename_columns = {
    'Q6_Personalisation options (eg: customisable avatars': 'Q6_Personalisation options',
    'Q6_Community and social features (eg: chat with friends': 'Q6_Community and social features',
    'Q6_Gamification (eg: challenges and quests': 'Q6_Gamification',
    'Q14_Health benefits (eg: fitness': 'Q14_Health benefits'
}
df_encoded_multiple.rename(columns=rename_columns, inplace=True)
```
